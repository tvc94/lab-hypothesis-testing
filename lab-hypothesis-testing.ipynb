- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.

```python

import scipy.stats as stats

# Separate data into Legendary and Non-Legendary
legendary = df[df['Legendary'] == True]
non_legendary = df[df['Legendary'] == False]

# List of stats to test
stats_to_test = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Dictionary to hold results
test_results = {}

# Perform t-test for each stat
for stat in stats_to_test:
    t_stat, p_value = stats.ttest_ind(legendary[stat], non_legendary[stat])
    test_results[stat] = (t_stat, p_value)

    print(f"{stat} - t-statistic: {t_stat}, p-value: {p_value}")

# Interpret results
for stat, (t_stat, p_value) in test_results.items():
    if p_value < 0.05:
        print(f"{stat}: Significant difference (p-value: {p_value})")
    else:
        print(f"{stat}: No significant difference (p-value: {p_value})")
```

it is true that legendary pokemon have different stats with non legendary pokemon on all stats

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

def calculate_euclidean_distances(df, school_coords, hospital_coords):
    """
    Calculate the Euclidean distance from each house to the school and hospital.

    Parameters:
    df (pd.DataFrame): DataFrame containing house coordinates with columns 'longitude' and 'latitude'.
    school_coords (tuple): Coordinates (longitude, latitude) of the school.
    hospital_coords (tuple): Coordinates (longitude, latitude) of the hospital.

    Returns:
    pd.DataFrame: DataFrame with two additional columns 'distance_to_school' and 'distance_to_hospital'.
    """
    df['distance_to_school'] = np.sqrt((df['longitude'] - school_coords[0]) ** 2 + (df['latitude'] - school_coords[1]) ** 2)
    df['distance_to_hospital'] = np.sqrt((df['longitude'] - hospital_coords[0]) ** 2 + (df['latitude'] - hospital_coords[1]) ** 2)
    return df

def categorize_houses(df, threshold):
    """
    Categorize houses based on their distance to the school and hospital.
    
    Parameters:
    df (pd.DataFrame): DataFrame containing distances with columns 'distance_to_school' and 'distance_to_hospital'.
    threshold (float): Distance threshold to categorize houses.
    
    Returns:
    pd.DataFrame: DataFrame with additional columns 'close_to_school' and 'close_to_hospital'.
    """
    df['close_to_school'] = df['distance_to_school'] < threshold
    df['close_to_hospital'] = df['distance_to_hospital'] < threshold
    return df

def perform_t_test(df, category, target_col='median_house_value'):
    """
    Perform t-test to compare close and far houses.

    Parameters:
    df (pd.DataFrame): DataFrame containing house data.
    category (str): The column to be tested, e.g., 'close_to_school' or 'close_to_hospital'.
    target_col (str): The column against which the t-test is performed, e.g., 'median_house_value'.

    Returns:
    dict: t-statistic and p-value
    """
    close_houses = df[df[category] == True][target_col]
    far_houses = df[df[category] == False][target_col]
    t_stat, p_value = stats.ttest_ind(close_houses, far_houses)
    return {'t_stat': t_stat, 'p_value': p_value}

# Define coordinates for the school and hospital
school_coords = (-118, 37)   # Example coordinates for the school
hospital_coords = (-122, 34) # Example coordinates for the hospital
threshold_dist = 0.5  # Set the threshold distance for categorizing houses

# Step 1: Calculate distances
df = calculate_euclidean_distances(df, school_coords, hospital_coords)

# Step 2: Categorize houses
df = categorize_houses(df, threshold_dist)

# Step 3: Perform t-tests
results_school = perform_t_test(df, 'close_to_school')
results_hospital = perform_t_test(df, 'close_to_hospital')

# Step 4 and 5: Interpret and print the results
significance_level = 0.05

print("T-test Results: School Proximity")
print(f"t-statistic: {results_school['t_stat']}, p-value: {results_school['p_value']}")
if results_school['p_value'] < significance_level:
    print("Significant difference in house prices between houses close to school and far from school.")
else:
    print("No significant difference in house prices between houses close to school and far from school.")

print("\nT-test Results: Hospital Proximity")
print(f"t-statistic: {results_hospital['t_stat']}, p-value: {results_hospital['p_value']}")
if results_hospital['p_value'] < significance_level:
    print("Significant difference in house prices between houses close to hospital and far from hospital.")
else:
    print("No significant difference in house prices between houses close to hospital and far from hospital.")
```
